Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

```python
def compute_pay(hours, rate):
  if hrs <= 40:
     pay = hrs * r
  else:
    rp = 40 * r
    op = (hrs - 40) * (r * 1.5)
    pay = rp + op

  return pay

hrs = int (input("Enter Hours: "))
r = float (input("Enter Rate: "))

tp = compute_pay(hrs, r)
print("Pay: ", tp)
```

**CHALLENGE**

```python
def computegrade(score):
    if score >= 97 and score <= 100:
        return '1.0'
    elif score >= 93 and score < 97:
        return '1.25'
    elif score >= 89 and score < 93:
        return '1.5'
    elif score >= 85 and score < 89:
        return '1.75'
    elif score >= 81 and score < 85:
        return '2.0'
    elif score >= 77 and score < 81:
        return '2.25'
    elif score >= 74 and score < 77:
        return '2.5'
    elif score >= 0 and score < 74:
        return '3.0'
    else:
        return 'Invalid score'


scores = [100, 97, 93, 89, 85, 81, 77, 74]
print("Score             Grade")
print("[ 97, 100]        1.0")
print("[ 93,  97)        1.25")
print("[ 89,  93)        1.5")
print("[ 85,  89)        1.75")
print("[ 81,  85)        2.0")
print("[ 77,  81)        2.25")
print("[ 74,  77)        2.5")
print("[ 0,   74)        3.0")


for score in scores:
    score = int (input("\nEnter your score: "))
    print("Your grade is: "+ computegrade(score))
```
